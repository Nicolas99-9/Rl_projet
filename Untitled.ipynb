```python
import numpy as np
import glob

def stats_file(filename):
    stats = []
    with open(filename) as f:
        content = f.readlines()
        current_run = []
        for line in content:
            line = line.strip()
            if "DEBUT ITERATION" in line:
                if len(current_run)>0:
                    stats.append(current_run)
                    current_run = []
            else:
                current_run.append(line.split(" "))
        stats.append(current_run)
    stats = np.array(stats).astype(float)
    #stats
    #np.max(stats),np.mean(stats),np.median(stats),stats[-1], reward_cuml , t, learning_rate
    print("---------------------------")
    print("filename " , filename)
    distance_max = np.max([np.max(stats[i][:,0])for i in range(len(stats))])
    print("Distance maximale parcourue (y val) : ",distance_max)
    distance_mean = np.mean([np.mean(stats[i][:,0])for i in range(len(stats))])
    print("Distance moyenne parcourue (y val) : ",distance_mean)
    distance_mean = ([np.mean(stats[i][-100:,1])for i in range(len(stats))])
    print("Distance moyenne parcourue (y val); 50 dernieres iterations : ",distance_mean, np.mean(distance_mean))
    distance_median= ([np.median(stats[i][:,1])for i in range(len(stats))])
    print("Distance mediane parcourue  : ",distance_median, np.mean(distance_median))
    moyenne_stop = ([np.mean(stats[i][-100:,3])for i in range(len(stats))])
    print("Distance moyenne d'arret  : ",moyenne_stop, np.mean(moyenne_stop))
    print("Moyenne / Median / Max reward cumule moyenne : ", np.mean([np.mean(stats[i][:,4])for i in range(len(stats))]),
    np.mean([np.median(stats[i][:,4])for i in range(len(stats))]),np.mean([np.max(stats[i][:,4])for i in range(len(stats))]))
    print("Moyenne / Median / Max reward cumule moyenne  50 dernieres: ", np.mean([np.mean(stats[i][-50:,4])for i in range(len(stats))]),
    np.mean([np.median(stats[i][-50:,4])for i in range(len(stats))]),np.mean([np.max(stats[i][-50:,4])for i in range(len(stats))]))#print("Nombre moyen / max de pas "  , np.mean(np.mean(stats[])))
    #nombre d'iterations
    print("Moyenne / Median / Max pas cumule moyenne : ", np.mean([np.mean(stats[i][:,5])for i in range(len(stats))]),
    np.mean([np.median(stats[i][:,5])for i in range(len(stats))]),np.mean([np.max(stats[i][:,5])for i in range(len(stats))]))
    print("Moyenne / Median / Max pas cumule moyenne  50 dernieres: ", np.mean([np.mean(stats[i][-50:,5])for i in range(len(stats))]),
    np.mean([np.median(stats[i][-50:,5])for i in range(len(stats))]),np.mean([np.max(stats[i][-50:,5])for i in range(len(stats))]))#print("Nombre moyen / max de pas "  , np.mean(np.mean(stats[])))
    return stats


tmp = glob.glob('res/copie_new/*.txt')
all_stats = []
for f in tmp:
    stats = stats_file(f)
    all_stats.append(np.mean(stats,axis=0))
all_stats = np.array(all_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn


def runningMeanFast (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'same')
    return sma

def show_fig(tab_evo,labels,title,means):
    fig, axes = plt.subplots(nrows=len(tab_evo))
    colors = ('#A577D8', '#DF8080', 'b')
    for ax, color,t,l,m in zip(axes, colors,tab_evo,labels,means):
        data = np.random.random(1) * np.random.random(10)
        ax.plot(t, color=color,label=l)
        ax.plot(runningMeanFast(t,m),color="b",label="rolling mean")
        ax.legend(loc="best")
    plt.show()
#evolution of the cumulated reward
print("cum reward")
print(" ".join([tmp[i] for i in [1,10]]))
tab = [all_stats[i][:,4] for i in [1,10]]
show_fig(tab,["RF 5 / Epsilon Greedy","RF 1 / Boltzmann"],"test",[5,20])
#distance parcourue
print("distance")
print(" ".join([tmp[i] for i in [1,10]]))
tab = [all_stats[i][:,0] for i in [1,10]]
show_fig(tab,["RF 5 / Epsilon Greedy","RF  1 / Boltzmann"],"test",[5,20])
#number of steps
print("steos")
to_test = [1,6]
print(" ".join([tmp[i] for i in to_test]))
tab = [all_stats[i][:,5] for i in to_test]
show_fig(tab,["RF 5 / Epsilon Greedy","RF 1 / Epsilon Greedy"],"test",[10,8])
```
